# 3. Create gym environment

```python
import gym
# import pybulletgym
from evaluation import TEST_EVAL
from actor_network import TEST_POLICY
from environment.env_filter import ENV
import pandas as pd



env = ENV(gym.make('pybullet_cartpole-v0'))

plane = p.loadURDF("/content/drive/My Drive/RL/pybullet_data/plane.urdf",[0, 0, -1])
env.reset()
yaw = 0
frames = []
for r in range(60):
  img = env.render(mode='rgb_array', _yaw = yaw)
  frames.append(img)
  yaw += 6

print("creating animated png, please about 10 seconds")
%time write_apng("/content/drive/My Drive/RL/results/example.png", frames, delay=100)
%time Image(filename="/content/drive/My Drive/RL/results/example.png")
```
